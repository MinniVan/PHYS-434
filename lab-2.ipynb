### 1) Gamma-rays

We are using a telescope and are looking for gamma-ray sources. However, there backgrounds of cosmic-rays providing a discrete noise that follows a Poisson distribution. We are to detect a gamma-ray source by asking what the probability that the cosmic-ray background would have given us a measurement as signal-like or more signal like than the signal we recieved.

We assume that in 1 day, the average cosmic-ray background is some number X (between 0.5 and 10); and the average number of gamma-rays emitted by our source is Y ( Y > X)

&nbsp;&nbsp;&nbsp; 1) Show how the probability distribution of the background changes as we integrate (sum) more days.

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
import scipy


```

```python
X = 5 # average cosmic-ray background 
Y = 7 #  average number of gamma-rays emitted by our source
num_events = np.arange(0, 16)
prob = poisson(X).pmf(num_events)
```

#### Day 1

```python


plt.rcParams['figure.figsize'] = [10, 5]
# DAY 1
fig, ax = plt.subplots(1, 1)
ax.step(num_events, prob, where = 'mid', label = "Cosmic Background") 
plt.title("Poisson Distribution of Cosmic Background (DAY 1)", size = 15)
plt.xlabel("Number of Events", size= 15)
plt.ylabel("Probability", size=15)

mu = X
def f(x):
   return np.exp(-mu)*(mu**x)/(scipy.special.factorial(x)) # Poisson distribution (the actual function)
x = np.linspace(0, 16, 1000)
ax.plot(x, f(x), color='red', label="Continuous pmf()")
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(loc="best")

plt.show()
```

#### Day 1 (semi-log)

```python
plt.rcParams['figure.figsize'] = [10, 5]
# DAY 1
fig, ax = plt.subplots(1, 1)
ax.step(num_events, prob, where = 'mid', label = "Cosmic Background") 
plt.yscale('log')
plt.title("Poisson Distribution of Cosmic Background (semi-logY) (DAY 1)", size = 15)
plt.xlabel("Number of Events", size= 15)
plt.ylabel("log(Probability)", size=15)


mu = X
def f(x):
   return np.exp(-mu)*(mu**x)/(scipy.special.factorial(x)) # Poisson distribution (the actual function)
x = np.linspace(0, 16, 1000)
ax.plot(x, f(x), color='red', label="Continuous pmf()")
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(loc="best")
plt.show()
```

#### Day 2 (one day is added)

```python
# day 2 (one day being added)

# convolve background (Poisson distribution) once with itself
background_day_2 = np.convolve(prob, prob, mode='full')
num_events2 = np.arange(0, 31)

plt.rcParams['figure.figsize'] = [10, 5]
fig, ax = plt.subplots(1, 1)
ax.step(num_events2, background_day_2, where = 'mid', label = "Cosmic Background") 
plt.title("Poisson Distribution of Cosmic Background (DAY 2)", size = 15)
plt.xlabel("Number of Events", size= 15)
plt.ylabel("Probability", size=15)


mu = 10 # because the day has advanced, average has increased (sum of two)
def f(x):
   return np.exp(-mu)*(mu**x)/(scipy.special.factorial(x)) # Poisson distribution (the actual function)
x = np.linspace(0, 30, 1000)

ax.plot(x, f(x), color='red', label="Continuous pmf()")
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(loc="best")
plt.show()
```
