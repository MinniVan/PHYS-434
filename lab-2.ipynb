## 1) Gamma-rays

We are using a telescope and are looking for gamma-ray sources. However, there backgrounds of cosmic-rays providing a discrete noise that follows a Poisson distribution. We are to detect a gamma-ray source by asking what the probability that the cosmic-ray background would have given us a measurement as signal-like or more signal like than the signal we recieved.

We assume that in 1 day, the average cosmic-ray background is some number X (between 0.5 and 10); and the average number of gamma-rays emitted by our source is Y ( Y > X)
<br/> Choosing X = 5, and Y = 7

### Part A

&nbsp;&nbsp;&nbsp; Show how the probability distribution of the background changes as we integrate (sum) more days.c

What we will find: 
<br/> As we can see as we added more days, the mean of the distribution is increased as the day goes by. The distribution becomes more sharp and the peak/highest probability (which is the mean of the distribution) decreases as time goes by. This is to ensure the integral evaluates to 1. 

```python
import numpy as np
import scipy
from scipy.stats import poisson
from scipy import stats
import matplotlib.pyplot as plt
import scipy


```

```python
X = 5 # average cosmic-ray background 
Y = 7 #  average number of gamma-rays emitted by our source
num_events = np.arange(0, 16)
prob = poisson(X).pmf(num_events)
```

#### Day 1

```python


plt.rcParams['figure.figsize'] = [10, 5]
# DAY 1
fig, ax = plt.subplots(1, 1)
ax.step(num_events, prob, where = 'mid', label = "Cosmic Background") 
plt.title("Poisson Distribution of Cosmic Background (DAY 1)", size = 15)
plt.xlabel("Number of Events", size= 15)
plt.ylabel("Probability", size=15)

mu = X
def f(x):
   return np.exp(-mu)*(mu**x)/(scipy.special.factorial(x)) # Poisson distribution (the actual function)
x = np.linspace(0, 16, 1000)
ax.plot(x, f(x), color='red', label="Continuous pmf()")
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(loc="best")

plt.show()
```

## 2) Background over a time interval (Rayleigh)
&nbsp;&nbsp;&nbsp;&nbsp; A Rayleigh distribution is a skewed and continuours background that describes our background over more observing intervals

### Part A

We want to how how the distribution changes as you average over more observing intervals

### &nbsp;&nbsp;&nbsp; The pdf() for a continuous Rayleigh Distribution: $ f(x, \sigma) = \frac{x}{\sigma^2} e^{\frac{-x^2}{2 \sigma^2}}$ where $\sigma$ is the scale parameter of the distribution. We want to show that the distribution changes as we average over more observing intervals

```python
plt.rcParams['figure.figsize'] = [10, 5]
fig, ax = plt.subplots(1, 1)
sigma = 2
def ray(x):
   return np.exp((-x**2)/(sigma**2))*(x/sigma**2)

x = np.linspace(0, 15, 1000)
ax.plot(x, ray(x), color='red', label="Continuous pdf()")
plt.title("Background distribution: Rayleigh Distribution ", size = 15)
plt.xlabel("Number of Events", size= 15)
plt.ylabel("Probability", size=15)
```

```python
#semi log
plt.rcParams['figure.figsize'] = [10, 5]
fig, ax = plt.subplots(1, 1)
plt.yscale('log')
ax.plot(x, ray(x), color='red', label="Continuous pdf()")
plt.title("Background distribution: Rayleigh Distribution (semi-logY) ", size = 15)
plt.xlabel("Number of Events", size= 15)
plt.ylabel("log(Probability)", size=15)
```

```python
ray_convolve = ray(x)
interval = 10

for i in range (interval-1):
    ray_convolve = np.convolve(ray(x), ray_convolve, mode = "full")
    
plt.rcParams['figure.figsize'] = [10, 5]
fig, ax = plt.subplots(1, 1)

x2 = np.linspace(0, 15/interval, 9991)
ax.plot(x2, ray_convolve, color='red', label="Continuous pdf()")
plt.title("Rayleigh Distribution " + str(interval) + " intervals", size = 15)
plt.xlabel("Number of Events", size= 15)
plt.ylabel("Probability", size=15)
plt.xlim(0, 0.6)
plt.show()
```
