```python
import numpy as np
import scipy
from scipy import stats
import matplotlib.pyplot as plt
```

## Problem 0: some fake data to mess with

First, let's make a time axis that goes from 0 to 100 Ï€  in 1e5 steps

```python
t = np.linspace(0,np.pi*100,int(1e5))
```

Then an offset sinusoid

```python
N = 10+np.sin(t)
```

Then some random background data whose mean and variance depend on N

```python
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

## Problem 1: subsets of data

1a) Make a scatter plot (not a line plot) of the first 1000 data points.

```python
colors = np.random.rand(1000)
plt.scatter(t[0:1000], bkgd[0:1000], c=colors)
plt.xlabel('Time')
plt.ylabel('Fake Fake')
plt.title('First 1000 data points')
plt.plot()
```

1b) Generalize your code so you can make a plot of any X contiguous points (e.g. 2000 points starting at point 500), and produce an example plot of a set of data somethere in the middle of your array.

```python

start = int(len(t)/4)
end = int(len(t)/2) + start

colors = np.random.rand(len(t[start:end]))
plt.scatter(t[start:end], bkgd[start:end], c=colors)
plt.xlabel('Time')
plt.ylabel('Fake Fake')
plt.title('Set of data in the middle')
plt.plot()

```

1c) Sampling data, plot every 100th points

```python
start = 0
end = len(bkgd)
step = 100

colors = np.random.rand(len(t[start:end:step]))
plt.scatter(t[start:end:step], bkgd[start:end:step], c=colors)
plt.xlabel('Time')
plt.ylabel('Fake Fake')
plt.title('Sampling every 100th data points')
plt.plot()
```
